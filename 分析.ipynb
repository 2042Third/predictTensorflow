```python
import numpy as np
import pandas as pd
from sklearn.multioutput import MultiOutputClassifier
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report, confusion_matrix, precision_score
import csv

fileName = "视频号直播历史数据.csv"
colNames=["直播主题","直播位置","直播开始时间","直播时长","观看人数"
          ,"观看次数","最高在线","平均观看时长","点赞次数","评论次数"
          ,"分享次数","新增关注","总热度","送礼人数","送礼次数"
          ,"成交金额","成交订单数","成交人数"]
csvFile=pd.read_csv(fileName, names=colNames, header=None)
```

```python
def splitNPart(str, key):
    tmp = str.split(key)
    if(len(tmp)>1):
        return [tmp[0],tmp[1]]
    else:
        return ["0",tmp[0]]
def parseTimeToSec(str):
    tmp = splitNPart(str,"小时")
    hr = int(tmp[0])
    tmp = splitNPart(tmp[1],"分钟")
    mt = int(tmp[0])
    tmp = splitNPart(tmp[1],"秒")
    sc = int(tmp[0])
    total = hr*60*60 + mt*60 + sc
    return total
    
    
```

```python
   
for idx, val in enumerate(csvFile["平均观看时长"]):
    csvFile.at[idx, "平均观看时长"] = parseTimeToSec(val)

trainxor = csvFile[['平均观看时长', "评论次数","分享次数"]]
trainyor = csvFile[['观看人数']]

print(trainxor)
```

```python
import tensorflow as tf

# Convert to float32
X_train = X_train.astype('float32')
y_train = y_train.astype('float32')
X_test = X_test.astype('float32')
y_test = y_test.astype('float32')

# # Split the data
# X_train, X_test, y_train, y_test = train_test_split(
#     trainxor, trainyor, test_size=0.1, random_state=0)

# Create a model
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(11, activation='relu', input_shape=(X_train.shape[1],)),
    tf.keras.layers.Dense(11, activation='relu'),
    tf.keras.layers.Dense(11, activation='relu'),
    tf.keras.layers.Dense(1)
])

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X_train, y_train, epochs=500)

# Make predictions
y_pred = model.predict(X_test)

# Print evaluation metrics
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
print("Mean Absolute Error:", mean_absolute_error(y_test, y_pred))
```
